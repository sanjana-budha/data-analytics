```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    "Student_ID": [101,102,103,104,105,106,107,108,109,110,111,112],
    "Marks": [95,75,88,91,89,66,60,87,65,92,89,90],
    "Study_Hours": [3,5,2,3,1,6,7,5,2,1,3,4]
}


df = pd.DataFrame(data)

print("Dataset:\n")
print(df)

print("\nStatistical Summary:\n")
print(df.describe())


print("\nMean Marks:", np.mean(df["Marks"]))
print("Mean Study Hours:", np.mean(df["Study_Hours"]))


print("\nMedian Marks:", np.median(df["Marks"]))
print("Median Study Hours:", np.median(df["Study_Hours"]))

print("\nMode Marks:", df["Marks"].mode()[0])
print("Mode Study Hours:", df["Study_Hours"].mode()[0])

print("\nStandard Deviation of Marks:", np.std(df["Marks"]))
print("Standard Deviation of Study Hours:", np.std(df["Study_Hours"]))


print("\nVariance of Marks:", np.var(df["Marks"]))
print("Variance of Study Hours:", np.var(df["Study_Hours"]))


correlation = df["Marks"].corr(df["Study_Hours"])
print("\nCorrelation between Marks and Study Hours:", correlation)


plt.scatter(df["Study_Hours"], df["Marks"])
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.show()

plt.hist(df["Marks"], bins=5)
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Distribution of Marks")
plt.show()

plt.boxplot(df["Marks"])
plt.title("Box Plot of Marks")
plt.show()
```
