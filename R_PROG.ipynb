```python
import statistics as stats
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# Dataset
data = [30, 40, 50, 60, 80, 90, 100]

print("Dataset:")
print(data)

# Mean
print("Mean =", np.mean(data))

# Median
print("Median =", np.median(data))

# Mode
counts = Counter(data)
max_freq = max(counts.values())

if max_freq == 1:
    print("Mode = No Mode")
else:
    mode_values = [k for k, v in counts.items() if v == max_freq]
    print("Mode =", mode_values)

# Maximum
print("Maximum =", max(data))

# Minimum
print("Minimum =", min(data))

# Variance
print("Variance =", np.var(data, ddof=1))

# Standard Deviation
print("Standard Deviation =", np.std(data, ddof=1))

# Summary
print("\nSummary:")
print("Count =", len(data))
print("Mean =", np.mean(data))
print("Min =", min(data))
print("Max =", max(data))

# Bar Plot
plt.figure()
plt.bar(range(1, len(data)+1), data)
plt.title("Bar Plot")
plt.xlabel("Observation")
plt.ylabel("Value")
plt.show()

# Histogram
plt.figure()
plt.hist(data)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Pie Chart
plt.figure()
plt.pie(data, labels=data, autopct="%1.1f%%")
plt.title("Pie Chart")
plt.show()
```
