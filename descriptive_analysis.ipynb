```python

#DESCRIPTIVE ANALYSIS
import statistics
import matplotlib.pyplot as plt

mem_name=["aman","riya","karan","sneha","rahul"]
ages=[22,25,24,23,26]
workout_hrs=[2,1.5,2.5,1,3]
calories_burned=[500,250,600,300,750]

print("--------Fitness center dataset------")
for i in range(len(mem_name)):
  print(mem_name[i],ages[i],workout_hrs[i],calories_burned[i])

avg_workout=sum(workout_hrs)/len(workout_hrs)
print("\nAverage Workout hours=",avg_workout)

max_calories=max(calories_burned)
print("\nMaximum calories burned=",max_calories)

min_calories=min(calories_burned)
print("\nMinimum calories burned=",min_calories)

median_calories=statistics.median(calories_burned)
print("\nMedian calories burned=",median_calories)

std_dev=statistics.stdev(calories_burned)
print("Standard Deviation=",std_dev)

index=calories_burned.index(max_calories)
print("Most Active Member=",mem_name[index])

plt.bar(mem_name,calories_burned)
plt.xlabel("Member Names")
plt.ylabel("Calories Burned")
plt.title("Fitness Analysis")
plt.show()

plt.pie(calories_burned,labels=mem_name)
plt.show()

plt.hist(calories_burned)
plt.show()

plt.scatter(mem_name,calories_burned)
plt.show()
```
