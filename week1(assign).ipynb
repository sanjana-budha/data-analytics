```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data={"oder_time":[10,2,10,4,6],
      "delivery_time":[10,50,30,35,12],
      "delivery_partner":["a","b","c","d","e"],
      "customer_ratings":[4,3,5,3,4]}

df=pd.DataFrame(data)
print(df)

#average delivery time
avg_delivery_time=df["delivery_time"].mean()
print("Average delivery time:",avg_delivery_time)

#peak oder hours
peak_oder_hours=df["oder_time"].mode()
print("Peak oder hours:\n",peak_oder_hours)

#most delayed areas
most_delayed_areas=df[df["delivery_time"]>avg_delivery_time]
print("Most delayed areas:\n",most_delayed_areas)

#top rated delivery patners
top_rated_delivery_patners=df.sort_values(by="customer_ratings",ascending=False).head(3)
print("Top rated delivery patners:\n",top_rated_delivery_patners)

plt.pie(df["customer_ratings"],labels=df["delivery_partner"],autopct="%1.1f%%")
plt.title("delivery ratings")
plt.show()
plt.bar(df["delivery_partner"],df["customer_ratings"])
plt.xlabel("Delivery partner")
plt.ylabel("customer_ratings")
plt.title("delivery ratings")
plt.show()

df.info()
df.describe()
```

```python
#customer buying behavior
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data={"product_purchase":["shoes","dress","dress","facewash","dryfruits"],
      #"cart_abandonment":[1,1,0,0,1],
      "cart_abandonment":["y","n","n","y","y"],
      "coustomer_ratings":[5,3,4,4,2],
      "payment_methods":["upi","upi","debit_card","cod","upi"]}

df=pd.DataFrame(data)
print(df.to_string(index=False))

#most sold products
most_sold_products=df["product_purchase"].mode()
print("Most sold products:",most_sold_products.to_string(index=False))

#high revenue coustomers
high_revenue_coustomers=df.sort_values(by="coustomer_ratings",ascending=False).head(3)
print("High revenue coustomers:\n",high_revenue_coustomers.to_string(index=False))

#most used payment method
most_used_payment_method=df["payment_methods"].mode()
print("Most used payment method:",most_used_payment_method.to_string(index=False))

#product with poor ratings
product_with_poor_ratings=df[df["coustomer_ratings"]<3]
print("Product with poor ratings:\n",product_with_poor_ratings.to_string(index=False))

plt.pie(df["coustomer_ratings"],labels=df["product_purchase"],autopct="%1.1f%%")
plt.title("coustomer ratings")
plt.show()
plt.bar(df["product_purchase"],df["coustomer_ratings"])
plt.xlabel("product_purchase")
plt.ylabel("coustomer_ratings")
plt.title("coustomer ratings")
plt.show()
```

```python
import pandas as pd

data = {
    "Product": ["Milk", "Bread", "Rice", "Eggs", "Juice"],
    "Daily_Sales": [50, 20, 5, 40, 2],
    "Product_Stock": [10, 50, 100, 15, 80],
    "Expiry_Days": [2, 5, 180, 7, 1]
}

df = pd.DataFrame(data)

print("Dataset")
print(df)

# Fast Moving Products
fast_moving = df[df["Daily_Sales"] > 30]

print("\nFast Moving Products")
print(fast_moving)

# Slow Moving Products
slow_moving = df[df["Daily_Sales"] < 10]

print("\nSlow Moving Products")
print(slow_moving)

# Out of Stock Trend
out_of_stock = df[df["Product_Stock"] < 20]

print("\nOut of Stock Trend")
print(out_of_stock)
# Products Near Expiry
near_expiry = df[df["Expiry_Days"] <= 3]

print("\nNear Expiry Products")
print(near_expiry)
```
