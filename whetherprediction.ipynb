```python
import pandas as pd
import matplotlib.pyplot as plt

# ---------------- DATASET ----------------
data = {
    "Days": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "Temperature_C": [28, 30, 29, 32, 34, 33, 35]
}

df = pd.DataFrame(data)

print("WEATHER DATASET")
print(df)

# ---------------- AVERAGE TEMPERATURE ----------------
avg = df["Temperature_C"].mean()
print("\nAverage Temperature =", avg)

# ---------------- TEMPERATURE INCREASE ----------------
increase = df["Temperature_C"].diff().fillna(0)

print("\nTemperature Increased =")
print(increase.to_string(index=False))

# ---------------- TOTAL INCREASE ----------------
sum_increase = increase.sum()
print("\nSum of Increased Temperature =", sum_increase)

# ---------------- PREDICTION ----------------
last_value = df["Temperature_C"].iloc[-1]

prediction = last_value + sum_increase

print("\nPredicted Future Temperature =", prediction)

# ---------------- DECISION ----------------
if prediction > avg:
    print("Weather Temperature Will Increase")
else:
    print("Weather Temperature Stable")

# ---------------- LINE GRAPH ----------------
plt.figure(figsize=(8, 5))

plt.plot(
    df["Days"],
    df["Temperature_C"],
    marker='o'
)

plt.xlabel("Days")
plt.ylabel("Temperature (°C)")
plt.title("Weather Temperature Prediction")

plt.grid(True)
plt.show()
```
